```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,color='yellow')
```

```python
sns.scatterplot(x="petal_length",y="petal_width",data=iris,color='pink')
```

```python
sns.scatterplot(iris["sepal_length"],iris["sepal_width"],data=iris,color='violet')
```

```python
sns.lineplot(iris["sepal_length"],iris["sepal_width"],data=iris,color='violet')
```

```python
sns.distplot(iris["sepal_length"],color='violet')
```

```python
sns.barplot(iris["petal_length"],iris["petal_width"],data=iris,color='yellow')
```

```python
sns.boxplot(iris["petal_length"],iris["petal_width"],data=iris)
```

```python
sns.catplot(x="sepal_length",y="species",data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr())
```

```python
sns.heatmap(iris.corr(),annot=True)
```
